```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('img1.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(image)
```

```python
# Applying bilateral filter to reduce noise while keeping edges sharp
blurred_image = cv2.bilateralFilter(gray_image, d=7, sigmaColor=50, sigmaSpace=50)
cv2_imshow(blurred_image)
```

```python
# Detecting edges using adaptive thresholding
edges = cv2.adaptiveThreshold(blurred_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 2)
```

```python
# Combining edges with the original image
cartoon_image = cv2.bitwise_and(image, image, mask=edges)
```

```python
cv2_imshow(cartoon_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# To save the image
cv2.imwrite('cartoonized_image.jpg', cartoon_image)
```
